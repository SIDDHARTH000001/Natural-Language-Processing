```python
import nltk
from nltk.corpus import brown

# Ensure the Brown Corpus is downloaded
nltk.download('brown')

# Define the days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Define genres of interest
news_genres = ['news']
romance_genres = ['romance']

# Create Conditional Frequency Distributions
cfd_news = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in news_genres
    for fileid in brown.fileids(categories=genre)
    for word in brown.words(fileid)
    if word in days
)

cfd_romance = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in romance_genres
    for fileid in brown.fileids(categories=genre)
    for word in brown.words(fileid)
    if word in days
)

# Tabulate counts for news and romance genres
print("News Genre Counts:")
cfd_news.tabulate(samples=days)

print("\nRomance Genre Counts:")
cfd_romance.tabulate(samples=days)

# Plot the counts
import matplotlib.pyplot as plt

cfd_news.plot(figsize=(10, 5),samples=days, title="Frequency of Days in News Genre")

cfd_romance.plot(figsize=(10, 5),samples=days, title="Frequency of Days in Romance Genre",color="orange")

```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```



## Let’s define a function to compute what fraction of words in a text are not in the stopwords list:


```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

content_fraction(nltk.corpus.reuters.words())
```

# Zipf’s Law

```python
import nltk
from nltk.corpus import reuters
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

nltk.download('reuters')
nltk.download('punkt')

# Load the text from the Reuters corpus
text = ' '.join(reuters.words())
```

```python
words = nltk.word_tokenize(text.lower())
freq_dist = nltk.FreqDist(words)

freqs = list(freq_dist.values())
ranks = np.arange(1, len(freqs) + 1)
```

```python
sorted_freqs = sorted(freqs, reverse=True)

plt.figure(figsize=(12, 6))
plt.plot(ranks, sorted_freqs, 'o-', label='Word Frequencies')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.title('Zipf\'s Law: Word Frequency vs. Rank')
plt.legend()
plt.grid(True)
plt.show()

print("Sample check for Zipf's Law:")
for i in range(1, 5):  # Check the first 4 words
    rank = i
    freq = sorted_freqs[i-1]  # Zero-based index
    predicted_freq = sorted_freqs[0] / rank
    print(f"Rank: {rank}, Actual Frequency: {freq}, Predicted Frequency: {predicted_freq:.2f}")
```

```python
import nltk
from nltk import word_tokenize
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

# Sample text
text = "New York City is known for its vibrant culture and iconic landmarks. New York is a city that never sleeps."

# Tokenize the text
tokens = word_tokenize(text.lower())

# Find bigrams
finder = BigramCollocationFinder.from_words(tokens)

# Apply a scoring method to find the best collocations
scorer = BigramAssocMeasures.likelihood_ratio
collocations = finder.nbest(scorer, 10)

collocations
```

```python
import nltk
from nltk.corpus import brown
from nltk.probability import FreqDist

# Load tagged words from the Brown corpus
brown_news_tagged = brown.tagged_words(categories='news')

# Optional: Define a mapping for simplifying tags if needed
simplify_tags = {
    'NN': 'NOUN', 'NNS': 'NOUN', 'NNP': 'NOUN', 'NNPS': 'NOUN',
    'VB': 'VERB', 'VBD': 'VERB', 'VBG': 'VERB', 'VBN': 'VERB', 'VBP': 'VERB', 'VBZ': 'VERB',
    'RB': 'ADV', 'RBR': 'ADV', 'RBS': 'ADV',
    'JJ': 'ADJ', 'JJR': 'ADJ', 'JJS': 'ADJ',
    'PRP': 'PRON', 'PRP$': 'PRON', 'WP': 'PRON', 'WP$': 'PRON',
    'IN': 'ADP', 'DT': 'DET', 'MD': 'AUX',
    'CC': 'CONJ', 'TO': 'PART', 'UH': 'INTJ', 'EX': 'EXCL'
}

# Simplify tags
simplified_brown_news_tagged = [(word, simplify_tags.get(tag, tag)) for (word, tag) in brown_news_tagged]

# Create a frequency distribution of the simplified tags
tag_fd = FreqDist(tag for (word, tag) in simplified_brown_news_tagged)
print(tag_fd.keys())
```

```python
def findtags(tag_prefix, tagged_text):
    cfd = nltk.ConditionalFreqDist((tag, word) for (word, tag) in tagged_text if tag.startswith(tag_prefix))
    return dict((tag, list(cfd[tag].keys())[:5]) for tag in cfd.conditions())

findtags('NN', nltk.corpus.brown.tagged_words(categories='news'))
```

```python
brown_tagged_sents = brown.tagged_sents(categories='news')
brown_sents = brown.sents(categories='news')

patterns = [
    (r'.*ing$', 'VBG'), # gerunds
    (r'.*ed$', 'VBD'), # simple past
    (r'.*es$', 'VBZ'), # 3rd singular present
    (r'.*ould$', 'MD'), # modals
    (r'.*\'s$', 'NN$'), # possessive nouns
    (r'.*s$', 'NNS'), # plural nouns
    (r'^-?[0-9]+(.[0-9]+)?$', 'CD'), # cardinal numbers
    (r'.*', 'NN') # nouns (default)
 ]

regexp_tagger = nltk.RegexpTagger(patterns)
regexp_tagger.tag(brown_sents[3])
```

```python
def performance(cfd, wordlist):
    lt = dict((word, cfd[word].max()) for word in wordlist)
    baseline_tagger = nltk.UnigramTagger(model=lt, backoff=nltk.DefaultTagger('NN'))
    return baseline_tagger.evaluate(brown.tagged_sents(categories='news'))
def display():
    import pylab
    words_by_freq = list(nltk.FreqDist(brown.words(categories='news')))
    cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))
    sizes = 2 ** pylab.arange(15)
    perfs = [performance(cfd, words_by_freq[:size]) for size in sizes]
    pylab.plot(sizes, perfs, '-bo')
    pylab.title('Lookup Tagger Performance with Varying Model Size')
    pylab.xlabel('Model Size')
    pylab.ylabel('Performance')
    pylab.show()
display() 
```
